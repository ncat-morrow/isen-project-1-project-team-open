Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 120)
```

Data Imports

```python
data_url = 'https://raw.githubusercontent.com/StevenXJiang/ISEN370/refs/heads/main/VU_Building_Data_for_ISEN370_Spr_2020.csv'
data = pd.read_csv(data_url)

data.head()
```

Exploring Data

```python
features = data[['Heating Water Flow Rate', 'Cooling Water Flow Rate', 'TotalEnergy']]
features.head()
```

Statistics

```python
stats = features.describe()
stats
```

Histograms

```python
plt.figure()
plt.hist(features['Heating Water Flow Rate'].dropna())
plt.title('Heating Water Flow Rate')
plt.xlabel('Heating Water Flow Rate')
plt.ylabel('Frequency')
plt.show()

plt.figure()
plt.hist(features['Cooling Water Flow Rate'].dropna())
plt.title('Cooling Water Flow Rate')
plt.xlabel('Cooling Water Flow Rate')
plt.ylabel('Frequency')
plt.show()

plt.figure()
plt.hist(features['TotalEnergy'].dropna())
plt.title('TotalEnergy')
plt.xlabel('TotalEnergy')
plt.ylabel('Frequency')
plt.show()
```

Scatter Plots vs Time

```python
# Convert time to datetime
data['Time_dt'] = pd.to_datetime(data['Time'], errors='coerce')

plt.figure()
plt.scatter(data['Time_dt'], data['Heating Water Flow Rate'], s=6)
plt.xlabel('Time'); plt.ylabel('Heating Water Flow Rate')
plt.title('Heating Water Flow Rate vs Time')
plt.xticks(rotation=25); plt.tight_layout(); plt.show()

plt.figure()
plt.scatter(data['Time_dt'], data['Cooling Water Flow Rate'], s=6)
plt.xlabel('Time'); plt.ylabel('Cooling Water Flow Rate')
plt.title('Cooling Water Flow Rate vs Time')
plt.xticks(rotation=25); plt.tight_layout(); plt.show()

plt.figure()
plt.scatter(data['Time_dt'], data['TotalEnergy'], s=6)
plt.xlabel('Time'); plt.ylabel('TotalEnergy')
plt.title('TotalEnergy vs Time')
plt.xticks(rotation=25); plt.tight_layout(); plt.show()
```

Distribution shapes:

Heating Water Flow Rate: The histogram is right-skewed. Most observations are at lower flow with some spikes to higher values, which makes sense because heating is not running hard all the time.

Cooling Water Flow Rate: Also right-skewed with a wider spread than heating. It looks like there are clusters that probably line up with warmer periods.

TotalEnergy: Strong right skew with a long tail. Most of the energy use sits in a moderate range, but there are notable peaks, likely times when both heating/cooling and other building loads are active.
